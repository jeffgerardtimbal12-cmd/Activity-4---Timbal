```python
from pathlib import Path
from datetime import datetime
import shutil

# Step 2: Define working directory
backup_dir = Path.home() / "Documents" / "PythonFiles"
backup_dir.mkdir(exist_ok=True, parents=True)

# Step 3: Write file with backup
def write_with_backup(filename: str, content: str):
    file_path = backup_dir / filename

    # Create backup if file exists
    if file_path.exists():
        timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
        backup_path = file_path.with_name(
            f"{file_path.stem}_backup_{timestamp}{file_path.suffix}"
        )
        shutil.copy2(file_path, backup_path)
        print(f"Backup saved: {backup_path.name}")

    # Write new content
    with open(file_path, "w", encoding="utf-8") as file:
        file.write(content)

    print(f"File saved: {file_path.name}")

# Step 4: Read file
def read_file(filename: str):
    file_path = backup_dir / filename
    with open(file_path, "r", encoding="utf-8") as file:
        return file.read()

# Step 5: List backups for a specific file
def list_backups(filename: str):
    stem = Path(filename).stem
    backups = list(backup_dir.glob(f"{stem}_backup_*"))
    return backups

# Step 6: Append or overwrite option
def save_file(filename: str, content: str, mode: str = "overwrite"):
    file_path = backup_dir / filename

    if mode == "append":
        with open(file_path, "a", encoding="utf-8") as file:
            file.write("\n" + content)
        print("Content appended.")
    else:
        write_with_backup(filename, content)

# Demo
print("=== File Operations Demo ===")

print("\n1. Creating new file:")
write_with_backup("demo.txt", "Initial content")

print("\n2. Updating file (with backup):")
write_with_backup("demo.txt", "Updated content")

print("\n3. Reading file:")
print(read_file("demo.txt"))

print("\n4. Listing backups:")
for backup in list_backups("demo.txt"):
    print("-", backup.name)
```

```python
from pathlib import Path
from datetime import datetime
import shutil

# 🔹 Change this to YOUR surname
SURNAME = "DelaCruz"

# Step 1: Working directory
backup_dir = Path.home() / "Documents" / f"{SURNAME}_Activity_5"
backup_dir.mkdir(exist_ok=True, parents=True)

# Step 2: File manager
def file_manager():
    file_name = input("Enter filename (e.g., notes.txt, data.csv, journal.md): ")
    file_path = backup_dir / file_name

    while True:
        print("\n--- MENU ---")
        print("1. Write to file")
        print("2. Append to file")
        print("3. Read file")
        print("4. Backup file")
        print("5. Exit")
        print("6. List backup files")

        choice = input("Choose an option (1-6): ")

        # 1. WRITE
        if choice == "1":
            content = input("Enter content to write:\n")
            with open(file_path, "w", encoding="utf-8") as f:
                f.write(content)
            print("File written successfully.")

        # 2. APPEND
        elif choice == "2":
            more = input("Enter content to append:\n")
            with open(file_path, "a", encoding="utf-8") as f:
                f.write("\n" + more)
            print("Content appended.")

        # 3. READ
        elif choice == "3":
            if file_path.exists():
                with open(file_path, "r", encoding="utf-8") as f:
                    print("\n--- FILE CONTENT ---\n")
                    print(f.read())
            else:
                print("File not found.")

        # 4. BACKUP (with surname in name)
        elif choice == "4":
            if file_path.exists():
                timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")

                backup_file = file_path.with_name(
                    f"{file_path.stem}_{SURNAME}_backup_{timestamp}{file_path.suffix}"
                )

                shutil.copy2(file_path, backup_file)
                print(f"Backup created: {backup_file.name}")
            else:
                print("Cannot backup. File does not exist.")

        # 5. EXIT
        elif choice == "5":
            print("Exiting file manager.")
            break

        # 6. LIST BACKUPS
        elif choice == "6":
            print("\n--- BACKUP FILES ---")
            backups = list(backup_dir.glob(f"*{SURNAME}_backup*"))

            if backups:
                for b in backups:
                    print("-", b.name)
            else:
                print("No backups found.")

        else:
            print("Invalid choice. Try again.")

# Step 3: Run program
if __name__ == "__main__":
    file_manager()
```

```python
from pathlib import Path

# Personal details
student_id = "2025-5535"

# File path (your setup)
documents_path = Path.home() / "Downloads" / "Activity_5_Files"

f1 = documents_path / f"intro_{student_id}.txt"
f2 = documents_path / f"lines_{student_id}.txt"
merged = documents_path / f"merged_{student_id}.txt"

# Ensure both files exist
if not f1.exists():
    f1.write_text(
        f"Welcome {student_id} to File Handling in Python!",
        encoding="utf-8"
    )
    print(f"Sample intro file created for Student ID: {student_id}")

if not f2.exists():
    f2.write_text("Line 1\nLine 2\nLine 3", encoding="utf-8")
    print(f"Sample lines file created for Student ID: {student_id}")

# Merge files
with merged.open("w", encoding="utf-8") as mf:
    mf.write(f1.read_text(encoding="utf-8"))
    mf.write("\n")
    mf.write(f2.read_text(encoding="utf-8"))

print(f"Files merged successfully for Student ID: {student_id}")
```
