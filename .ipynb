```python
# Step 1: Define an algorithm in pseudocode as comments
# Algorithm to find the largest number in a list
# 1. Assume the first number is the largest
# 2. Compare each number with the current largest
# 3. If a number is larger, update the largest
# 4. Return the largest number
# Step 2: Implement it in Python

numbers = [10, 5, 20, 8, 15]

Largest = numbers[0]

for num in numbers:
    if num > Largest:
        Largest = num

print("Largest number:", Largest)
```

```python
# Smallest Number
numbers = [10, 5, 20, 8, 15]

smallest = numbers[0]

for num in numbers:
    if num < smallest:
        smallest = num

print("Smallest number:", smallest)
```

```python
# Step 1: Implement bubble sort
def bubble_sort(arr):
    n = len(arr)
    
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                # swap
                arr[j], arr[j + 1] = arr[j + 1], arr[j]

# Step 2: Test sorting
to_sort = [64, 34, 25, 12, 22, 11, 90]
bubble_sort(to_sort)
print("Sorted list:", to_sort)
```

```python
# Number of Swaps 
def bubble_sort(arr):
    n = len(arr)
    swaps = 0  # counter for swaps
    
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                # swap
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swaps += 1
    
    return swaps

# Test the function
to_sort = [64, 34, 25, 12, 22, 11, 90]

swap_count = bubble_sort(to_sort)

print("Sorted list:", to_sort)
print("Total swaps:", swap_count)
```

```python
# Step 1: Implement selection sort
def selection_sort(arr):
    for i in range(len(arr)):
        min_idx = i
        
        for j in range(i + 1, len(arr)):
            if arr[j] < arr[min_idx]:
                min_idx = j
        
        # swap after finding smallest
        arr[i], arr[min_idx] = arr[min_idx], arr[i]

# Step 2: Test and observe performance
sample_data = [29, 10, 14, 37, 13]
selection_sort(sample_data)

print("Sorted list:", sample_data)
```

```python
# Number of comparison and swaps in selection sort vs bubble sort
def bubble_sort(arr):
    arr = arr.copy()
    n = len(arr)
    comparisons = 0
    swaps = 0

    for i in range(n):
        for j in range(0, n - i - 1):
            comparisons += 1
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swaps += 1

    return comparisons, swaps


def selection_sort(arr):
    arr = arr.copy()
    n = len(arr)
    comparisons = 0
    swaps = 0

    for i in range(n):
        min_idx = i

        for j in range(i + 1, n):
            comparisons += 1
            if arr[j] < arr[min_idx]:
                min_idx = j

        # swap only if needed
        if min_idx != i:
            arr[i], arr[min_idx] = arr[min_idx], arr[i]
            swaps += 1

    return comparisons, swaps


# Test data
data = [29, 10, 14, 37, 13]

bubble_comp, bubble_swaps = bubble_sort(data)
select_comp, select_swaps = selection_sort(data)

print("Bubble Sort:")
print("Comparisons:", bubble_comp)
print("Swaps:", bubble_swaps)

print("\nSelection Sort:")
print("Comparisons:", select_comp)
print("Swaps:", select_swaps)
```

```python
# Step 1: Implement insertion sort
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1

        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1

        arr[j + 1] = key

# Step 2: Test and measure performance
arr = [12, 11, 13, 5, 6]
insertion_sort(arr)
print("Sorted array:", arr)

```

```python
def fib_iter(n):
    a, b = 0, 1
    for _ in range(n):
        a, b = b, a + b
    return a
print(fib_iter(10))
```

```python
def fib_rec(n):
    if n <= 1:
        return n
    return fib_rec(n-1) + fib_rec(n-2)
print(fib_rec(10))
```

```python
def selection_sort(arr):
    comparisons, swaps = 0, 0
    for i in range(len(arr)):
        min_idx = i
        for j in range(i+1, len(arr)):
            comparisons += 1
            if arr[j] < arr[min_idx]:
                min_idx = j
        swaps += 1
        arr[i], arr[min_idx] = arr[min_idx], arr[i] 
    return arr, comparisons, swaps
print(selection_sort([5, 3, 8, 4, 2]))
```

```python
def insertion_sort(arr):
    comparisons, swaps = 0, 0

    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1

        while j >= 0 and arr[j] > key:
            comparisons += 1
            arr[j + 1] = arr[j]
            swaps += 1
            j -= 1

        arr[j + 1] = key

    return arr, comparisons, swaps


print(insertion_sort([5, 3, 8, 4, 2]))
```
